# Multivarate case

```python
monkey_data=  np.random.multivariate_normal([1,2],[[1,0.5],[0.5,1]],1000)
chimp_data=  np.random.multivariate_normal([4,4],[[1,0.5],[0.5,1]],1000)

print(monkey_data.shape)
```

```python
plt.scatter(monkey_data[:,0],monkey_data[:,1],label='Monkey')
plt.scatter(chimp_data[:,0],chimp_data[:,1],alpha=0.5,label='Chimpanzee')


query_x = np.array([3,2])
plt.scatter(query_x[0],query_x[1],c='k',label='Query Point')
plt.legend()
plt.show()
```

#K-Nearest Nieghbours 
 

```python
data = np.zeros((2000,3))

data[:1000,:-1] = monkey_data
data[1000:,:-1] = chimp_data
data[:1000,-1] = 0
data[1000:,-1] = 1


print(data.shape)
```

```python
x_train = data[:,:-1]
y_train = data[:,-1]
print(x_train.shape)
print(y_train.shape)
```

```python
def dist(x1,x2):
    return np.sqrt(((x1-x2)**2).sum())


def knn(x_train,y_train,query_x,k=5):
    dist_vals=[]
    m=x_train.shape[0]
    for ix in range(m):
        d=dist(query_x,x_train[ix])
        dist_vals.append((d,y_train[ix]))
    dist_vals = sorted(dist_vals)
    dist_vals=dist_vals[:k]
    
    y=np.array(dist_vals)
    t=np.unique(y[:,1],return_counts=True)
    index=t[1].argmax()
    prediction=t[0][index]
    
    print (prediction)
    
```

```python
knn(x_train,y_train,query_x)
```
